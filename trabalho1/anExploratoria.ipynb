### Importação das Bibliotecas Externas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("train.csv")
data.head()
```

### Análise Exploratória dos Dados

```python
labels = list(data.columns[2:])
labels.remove('Flight Distance')
labels.remove('Arrival Delay in Minutes')
labels.remove('Departure Delay in Minutes')
print(labels)
```

```python
for label in labels:
    plt.figure(figsize=[20,10])
    sns.countplot(x=label, data=data, palette="flare")
    _ = plt.ylabel(label)
    _ = plt.show()
```

```python
labels2 = ['Flight Distance', 'Departure Delay in Minutes', 'Arrival Delay in Minutes']
for label in labels2:
    plt.figure(figsize= [20,10])
    sns.kdeplot(data[label], color = 'red', fill = True)
    plt.xlabel(label)
    _ = plt.ylabel('Total')
    _ = plt.show()
```
